PROGRAM OBJECTIVE:

Prove that for semigroups of order 6, four filled cells of Cayley table
 are insufficient for unique determination.

This program do 2st part of proof:

3) Import 42 tables satisfying constrain
T[0][0]=1, T[1][1]=2, T[1][2]=0,

4) Showing that adding 4th filled cell gets to us no less than 2 satisfying tables

```python
from itertools import permutations
n=6
data=list(range(n))
Sn=list(permutations(data))
TableSet=set()
```

```python
# Checking if Sn is symmetric group order n
print(Sn)
```

```python
def compare(a,b,s):
    for i in range(n):
        for j in range(n):
            if a[i][j] != b[s[i]][s[j]]:
                return False
    return True

def isIsomorf(a, b, constrain):
    for s in Sn:
        if compare (a, b, s):
            return True
    return False
```

PHASE 3: Import 42 tables satisfying constrai

T[0][0]=1, T[1][1]=2, T[1][2]=0

```python
with open(file_semigroups, 'r', encoding='utf-8') as file:
    count = 0
    table_list = []
    table=[]
    for line in file:
        count +=1
        array = line[1:-2].split(', ') # Reading the original multiplication table
        if count%7==0:
            table=[]
        else:
            addition = [int(x) for x in array]
            table.append(addition)
            if count%7==6:
                table_list.append(table)

print('The number of tables satisfying the 3-cell ocnstrain is ', len(table_list))
```

PHASE 4:

We take 42 order-6 semigroups satisfying constrain

T[0][0]=1, T[1][1]=2, T[1][2]=0

and show that adding of 4th filled cell do not determinate a table uniquely

```python
# The 4th constraint is T[r1][c1]=v1
sati = 0
sati_max = 3
for r1 in range(n):
    for c1 in range(n):
        for v1 in range(n):
            sati = 0
            sati_tables = []
            for tt in table_list:
                if tt[r1][c1]==v1:
                    sati +=1
                    sati_tables.append(tt)
            if 0<sati<sati_max:
                print()
                print('The 4th constrain [ ', r1, c1, v1, ']')
                print('give us 2 satisfying tables')
                for t0 in sati_tables[0]:
                    print(*t0)
                print()
                for t1 in sati_tables[1]:
                    print(*t1)
                if isIsomorf(sati_tables[0], sati_tables[1], [r1,c1,v1]):
                    print('They are isomorfic')
                else:
                    print('They are not isomorfic')
                sati_max=2
```
